```python
# Task 2: Multi Input Prompt System

from langchain_core.prompts import PromptTemplate

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Case 1
output1 = multi_prompt.format(
    topic="AI",
    audience="beginners",
    tone="friendly"
)

# Test Case 2
output2 = multi_prompt.format(
    topic="Python",
    audience="kids",
    tone="fun"
)

# Test Case 3
output3 = multi_prompt.format(
    topic="Deep Learning",
    audience="engineers",
    tone="technical"
)

print(output1)
print()
print(output2)
print()
print(output3)
```

```python
# Task 3: Prompt Variations Engine

from langchain_core.prompts import PromptTemplate

# Teaching Prompt
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Prompt
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Prompt
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic_name = "Machine Learning"

print("Teaching Prompt:")
print(teaching_prompt.format(topic=topic_name))

print("\nInterview Prompt:")
print(interview_prompt.format(topic=topic_name))

print("\nStorytelling Prompt:")
print(story_prompt.format(topic=topic_name))
```

```python
# ChatPromptTemplate System
```

```python
# Task 4: ChatPromptTemplate System

from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role}."),
    ("human", "Explain {topic}")
])

messages = chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)

for message in messages:
    print(message)
```

```python
# Input Validation Layer
```

```python
# Task 5: Input Validation Layer

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Audience Validation
    if audience not in valid_audience:
        print("Invalid audience! Default value assigned.")
        audience = "beginner"

    # Tone Validation
    if tone not in valid_tone:
        print("Invalid tone! Default value assigned.")
        tone = "formal"

    return audience, tone


# Testing Validation Function
audience, tone = validate_inputs("kids", "friendly")

print("Validated Audience:", audience)
print("Validated Tone:", tone)
```

```python
# Prompt Generator App
```

```python
# Task 6: Prompt Generator App

from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):

    # Validate Inputs
    audience, tone = validate_inputs(audience, tone)

    # Prompt Template
    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style"
    )

    # Generate Prompt
    final_prompt = prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

    return final_prompt


# Test Function
result = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print(result)
```

```python
# Task 7: Template Reusability Test

from langchain_core.prompts import PromptTemplate

reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple language"
)

topics = [
    "Artificial Intelligence",
    "Python",
    "Machine Learning",
    "Deep Learning",
    "Data Science"
]

for topic in topics:
    print(reuse_template.format(topic=topic))
    print()
```
